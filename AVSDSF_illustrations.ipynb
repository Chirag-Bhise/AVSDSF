### **1. Evaluation of AV services handling success rate**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = np.array([
    [95, 97, 94, 90],   # AVERSER-SA
    [66, 70, 58, 55],   # PAGURUS
    [45, 50, 43, 38],   # LDLS
    [26, 24, 20, 18],   # ONCO
])

algorithms = ['AVSDSF', 'PAGURUS', 'LDLS', 'ONCO']
tree_sizes = [25, 50, 75, 100]

df = pd.DataFrame(data, index=algorithms, columns=tree_sizes)

plt.figure(figsize=(10, 6))
sns.heatmap(
    df,
    annot=True,
    cmap="YlGnBu",
    linewidths=.5,
    cbar_kws={'label': 'Success Rate (%)'},
    annot_kws={"size": 12}
)
plt.title("Heatmap of Success Rate of Prefetching", fontsize=16)
plt.xlabel("Size of AV application tree (n)", fontsize=14)
plt.ylabel("Approach", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

### **2. Evaluation of AVs prioritized services handling success rate**

```python
import matplotlib.pyplot as plt
import numpy as np

# Dataset
data = {
    'AVSDSF': {
        'Normal': [75, 72, 80, 81],
        'High':   [95, 90, 70, 95]
    },
    'ONCO': {
        'Normal': [92, 89, 68, 75],
        'High':   [90, 80, 65, 65]
    },
    'PAGURUS': {
        'Normal': [78, 52, 52, 50],
        'High':   [79, 62, 70, 70]
    },
    'LDLS': {
        'Normal': [86, 82, 82, 84],
        'High':   [80, 80, 81, 83]
    }
}

# X-axis: Number of RSUs
rsus = [25, 50, 75, 100]
x = np.array(rsus)

# Maximally distinct and darker colors
color_map = {
    'AVSDSF': '#004488',  # Dark Blue
    'ONCO': '#BB0000',    # Dark Red
}

# Line styles for conditions
linestyles = {
    'Normal': '-',
    'High': '--'
}

# Create the plot
plt.figure(figsize=(10, 6))

# Plotting with improved visibility
for algo in ['AVSDSF', 'ONCO']:
    for condition in ['Normal', 'High']:
        plt.plot(
            x,
            data[algo][condition],
            label=f"{algo} - {condition}",
            color=color_map[algo],
            linestyle=linestyles[condition],
            linewidth=3,
            marker='o' if condition == 'Normal' else 's',
            markersize=8
        )

# Enhanced title and labels with increased font sizes
plt.title("Success Rate Comparison: AVSDSF vs ONCO (N = 100)", fontsize=18, fontweight='bold')
plt.xlabel("Number of RSUs (K)", fontsize=16)
plt.ylabel("Success Rate (%)", fontsize=16)
plt.xticks(x, fontsize=14)
plt.yticks(fontsize=14)

# Improved grid and legend with larger font size
plt.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)
plt.legend(fontsize=14)
plt.tight_layout()
plt.show()
```

## **4. Evaluation of optimal RSU utilization**

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
labels = [r'$\lambda$ = 50', r'$\lambda$ = 40', r'$\lambda$ = 30', r'$\lambda$ = 20', r'$\lambda$ = 10']
ONCO =    [90.28, 80.00, 60.00, 40.00, 20.00]
PBO =     [49.43, 46.19, 44.80, 42.15, 52.70]
PAGURUS = [50.75, 50.54, 49.92, 48.07, 49.02]
MMTO =    [97.44, 97.42, 75.10, 38.72, 15.00]
LDLS =    [70.84, 70.78, 70.67, 70.59, 70.42]
AVSDSF =  [98.31, 98.30, 98.25, 77.79, 40.90]

# Stack the data for each category
data = np.array([ONCO, PBO, PAGURUS, MMTO, LDLS, AVSDSF])
category_names = ['ONCO', 'PBO', 'PAGURUS', 'MMTO', 'LDLS', 'AVSDSF']
colors = ['#1f77b4', '#ff7f0e', '#ffeb3b', '#2ca02c', '#e377c2', '#87cefa']  # matched to legend

# Position of bars on y-axis
y = np.arange(len(labels))

fig, ax = plt.subplots(figsize=(10, 5))

# Plot each stacked segment
left = np.zeros(len(labels))
for idx, (category, color) in enumerate(zip(category_names, colors)):
    ax.barh(y, data[idx], left=left, label=category, color=color)
    for i, (value, l) in enumerate(zip(data[idx], left)):
        ax.text(l + value / 2, i, f'{value:.2f}', va='center', ha='center', fontsize=8)
    left += data[idx]

# Labels and Titles
ax.set_xlabel('Edge Resource Utilization (%)')
ax.set_title('ERU across different service arrival rates')
ax.set_yticks(y)
ax.set_yticklabels(labels)
ax.legend(loc='lower right', bbox_to_anchor=(1.1, 0))

plt.tight_layout()
plt.show()
```
